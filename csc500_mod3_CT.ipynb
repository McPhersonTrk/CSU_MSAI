```python
##### Module 3 Critical Thinking
##### Colorado State University Global Campus
##### csc500-1
##### Dr Jessica Schwartz 

# Part 1.
"""Write a program that calculates the total amount of a meal purchased at a restaurant. 
The program should ask the user to enter the charge for the food and then calculate the amounts with an 18 percent tip and 7 percent sales tax. 
Display each of these amounts and the total price."""

# The program should ask the user to enter the number of each item ordered and then calculate the subtotal, tax, tip, and total.
# The tax is 7% and the tip is 18%.

def main():
    # Display the menu
    print("Welcome to the Food Truck!\nMenu\n1. Hamburger: $6.00\n2. Cheeseburger: $6.50\n3. Hot Dog: $3.50\n4. Chili Dog: $4.00\n5. Fries: $2.00\n6. Onion Rings: $2.50\n7. Soda: $1.50\n8. Water: $1.00")
    menu = {1: "hamburger", 2: "cheeseburger", 3: "hot Dog", 4: "chili dog", 5: "fries", 6: "onion oings", 7: "soda", 8: "water"}
    order= {}
    while True:
        item= int(input("Enter the menu item number for the item you'd like: "))
        quantity = int(input(f"How many {item}'s would you like? "))
        order[item] =order.get(item, 0) + quantity
        more = input("Would you like to order more? (y/n)").strip().lower()
        if more == "n":
            break
    print("\nYour order:")
    for item, quantity in order.items():
        print(f"{menu[item]}: {quantity}")
    # get the number of each item ordered
    hamburger, cheeseburger, hot_dog, chili_dog, fries, onion_rings, soda, water = [order.get(i, 0) for i in range(1, 9)]
    # Calculate the subtotal
    subtotal = hamburger * 6.00 + cheeseburger * 6.50 + hot_dog * 3.50 + chili_dog * 4.00 + fries * 2.00 + onion_rings * 2.50 + soda * 1.50 + water * 1.00
    # Calculate the tax
    tax = subtotal * 0.07
    # Calculate the tip
    tip = subtotal * 0.18
    # Calculate the total
    total = subtotal + tax + tip
    # Display the subtotal, tax, tip, and total
    print("Subtotal: ${:.2f}".format(subtotal))
    print("Tax: ${:.2f}".format(tax))
    print("Tip: ${:.2f}".format(tip))
    print("Total: ${:.2f}".format(total))

# Call the main function
main()
```

```python
#Part 2:
'''Many people keep time using a 24-hour clock (11 is 11am and 23 is 11pm, 0 is midnight). 
If it is currently 13 and you set your alarm to go off in 50 hours, it will be 15 (3pm). 
Write a Python program to solve the general version of the above problem. 
Ask the user for the time now (in hours) and then ask for the number of hours to wait for the alarm. 
Your program should output what the time will be on a 24-hour clock when the alarm goes off.'''

from datetime import datetime, timedelta

# ask user if they want to use the current time or set a start time
current_time = input("Would you like to use the current time? (y/n): ").strip().lower()
if current_time == "y":
    start_time = datetime.now()
else:
    start_time = datetime.strptime(input("Enter the start time in 24-hour format (HH:MM): "), "%H:%M")

# get the number of hours to wait for the alarm
hours = float(input("Enter the amount of hours to wait for the alarm: "))
# calculate the future time
future_time = start_time + timedelta(hours=hours)
# display the future time
print("The alarm will go off on: {}".format(future_time.strftime("%Y-%B-%d at %H:%M:%S")))
```
