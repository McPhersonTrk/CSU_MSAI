```python
#==================================================================================================
# Module 4 Portfolio Project - Shopping Cart
##==================================================================================================
# PART 1:
# Define the ItemToPurchase class, which models an item with a name, price, and quantity
class ItemToPurchase:
    # Constructor: Initializes item attributes with default values
    def __init__(self, item_name = "none", item_price = 0, item_quantity = 0):
        #attributes
        self.item_name = item_name
        self.item_price = item_price
        self.item_quantity = item_quantity
    # Method to print the cost of an individual item (qty * price).
    def print_item_cost(self):
        total_cost = self.item_price * self.item_quantity
        # Output and format the item name, quantity, price and total cost
        print(f'{self.item_name} {self.item_quantity} @ ${self.item_price:.2f} = ${total_cost:.2f}')
# PART 2:        
    # Method to return the total cost of an item used for total cost calculations.
    def get_total_cost(self):
        return self.item_price * self.item_quantity
# define a main function for the user to input the item name, price and quantity
def main():
    # prompt user to enter input for item_1
    item_1_name = input('Enter the item name:\n') # get name of item_1
    item_1_price = float(input('Enter the item price:\n')) # get price of item_1
    item_1_quantity = int(input('Enter the item quantity:\n')) # get quantity of item_1
    # create an ItemToPurchase object for Item_1 using the inputed data.
    item_1 = ItemToPurchase(item_1_name, item_1_price, item_1_quantity)
    #print(f'\n{item_1_name}')
    # prompt user to enter input foritem_2
    item_2_name = input('Enter the item name:\n')# get name of item_2
    item_2_price = float(input('Enter the item price:\n')) # get price of item_2
    item_2_quantity = int(input('Enter the item quantity:\n')) # get quantity of item_2
    # create an ItemToPurchase object for Item_2 using the inputed data
    item_2 = ItemToPurchase(item_2_name, item_2_price, item_2_quantity)
    #print(f'\n{item_2_name}')
# PART 3
    # print the output for quantity, price and total cost of item_1 and item_2, respectively
    print('\nTOTAL COST') # label for total cost section
    item_1.print_item_cost() #prints cost details for item_1
    item_2.print_item_cost() #prints cost details for item_2
    # calculate total cost for all items.
    total_cost = item_1.get_total_cost() + item_2.get_total_cost() # total cost of all items
    print(f'\nTotal: ${total_cost:.2f}')# print the total cost of all items

# call the main function to start the program.
if __name__ == "__main__":
    main()
##==================================================================================================
# Module 6 Portfolio Project - Shopping Cart    
##==================================================================================================
#Part #4
#define Shoppingcart class
class ShoppingCart:
    #parameterized constructor
    def __init__(self, customer_name = "none", current_date = "January 1, 2020"):
        #attributes
        self.customer_name = customer_name
        self.current_date = current_date
        self.cart_items = []
    #method to add items to cart
    def add_item(self, item):
        self.cart_items.append(item)
    #method to remove items from cart
    def remove_item(self, item_name):
        found=False
        for item in self.cart_items:
            if item.item_name == item_name:
                self.cart_items.remove(item)
                found=True
                break
        if not found:
            print('Item not found in cart.')
    #method to modify items in cart
    def modify_item(self, item_modify):
        found=False
        for item in self.cart_items:
            if item.item_name == item_modify.item_name:
                found=True
                if item_modify.item_price != 0:
                    item.item_price = item_modify.item_price
                if item_modify.item_quantity != 0:
                    item.item_quantity = item_modify.item_quantity
                break
        if not found:           
            print('Item not found in cart.')
    #method to get number of items in cart
    def get_items_in_cart(self):
        total_qty= sum(item.item_quanty for item in self.cart_items)
        return total_qty
    #method to get cost of items in cart
    def get_cost_of_cart(self):
        total_cost = sum(item.item_price * item.item_quantity for item in self.cart_items)
        return total_cost
    #method to print total cost of items in cart    
    def print_total_cost(self):
            print(f"{self.customer_name}'s Shopping Cart - {self.current_date}" )
            total_items = self.get_items_in_cart()
            print(f"number of items: {total_items}")
            if total_items == 0:
                print('Cart is Empty')
            else:
                for item in self.cart_items:
                    print(f"Totaol: ${self.get_cost_of_cart():.2f}")
    #method to print descriptions of items in cart
    def print_description(self):
        print(f"{self.customer_name}'s Cart: {self.current_date}    ")
        print('Item descriptions')
        for item in self.cart_items:
            print(f"{item.item_name}: {item.item_description}")
            
#part 5
# Fubction to display the shopping cart menu
def display_menu(cart):
    option = " "
    while option != "q":
        print('\nMENU')
        print('a - Add item to cart')
        print('r - Remove item from cart')
        print('c - Change item quantity')
        print('i - Output items\' descriptions')
        print('o - Output shopping cart')
        print('q - Quit')
        option = input('Choose an option:\n')
# option a: add item to cart
        if option == "a":
            item_name= input('Enter the item name:\n')
            item_price = float(input('Enter the item price:\n'))
            item_quantity = int(input('Enter the item quantity:\n'))
            new_item= ItemToPurchase(item_name, item_price, item_quantity)
            cart.add_item(new_item)
# option r: remove item from cart
        elif option == "r":
            item_name = input('Enter the name of the item to remove:\n')
            cart.remove_item(item_name)
# option c: change item quantity        
        elif option == "c":
            item_name = input('Enter the name of the item to modify:\n')
            item_price = float(input('Enter the new price:\n'))
            item_quantity = int(input('Enter the new quantity:\n'))
            new_item = ItemToPurchase(item_name, item_price, item_quantity)
            cart.modify_item(new_item)
# option i: output items' descriptions        
        elif option == "i":
            cart.print_description()
# option o: output shopping cart        
        elif option == "o":
            cart.print_total_cost()
# option q: quit the program
        elif option == "q":
            break
        else:
            print('Invalid option')
#part 6
# create a new shopping cart object
if __name__ == "__main__":
    # create a new shopping cart object
    cart = ShoppingCart()
    # call the display menu function to display the menu
    display_menu(cart)

##==================================================================================================
# Module 8 Portfolio Project - Shopping Cart
##==================================================================================================
```
