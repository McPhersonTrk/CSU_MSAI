```python
from datetime import datetime, timedelta

# Course start date
start_date = datetime(2024, 9, 9)

# Function to calculate due dates based on start date and week day
def calculate_due_date(week_number, day_name):
    week_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    start_week_day = start_date.weekday()
    day_index = week_days.index(day_name)
    
    # Calculate the number of days to add to the start date
    days_to_add = (week_number - 1) * 7 + (day_index - start_week_day)
    return start_date + timedelta(days=days_to_add)

# Assignments data
assignments = [
    {"name": "BioSig-ID™ validation - Module 3: Critical Thinking", "module": 3, "type": "BioSig-ID™", "due_day": "Wednesday"},
    {"name": "BioSig-ID™ validation - Module 4: Critical Thinking", "module": 4, "type": "BioSig-ID™", "due_day": "Wednesday"},
    {"name": "BioSig-ID™ validation - Module 6: Critical Thinking", "module": 6, "type": "BioSig-ID™", "due_day": "Wednesday"},
    {"name": "Module 1: Critical Thinking", "module": 1, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 1: Discussion Forum", "module": 1, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 2: Critical Thinking", "module": 2, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 2: Discussion Forum", "module": 2, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 3: Critical Thinking", "module": 3, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 3: Discussion Forum", "module": 3, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 4: Critical Thinking", "module": 4, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 4: Discussion Forum", "module": 4, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 4: Portfolio Milestone", "module": 4, "type": "Portfolio Milestone", "due_day": "Sunday"},
    {"name": "Module 5: Critical Thinking", "module": 5, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 5: Discussion Forum", "module": 5, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 6: Critical Thinking", "module": 6, "type": "Critical Thinking", "due_day": "Sunday"},
    {"name": "Module 6: Discussion Forum", "module": 6, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 7: Discussion Forum", "module": 7, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 8: Discussion Forum", "module": 8, "type": "Discussion", "due_day": "Wednesday"},
    {"name": "Module 8: Portfolio Project", "module": 8, "type": "Portfolio Project", "due_day": "Sunday"},
]

# Generating the formatted assignment list with due dates and sorting them
formatted_assignments = []
for assignment in assignments:
    due_date = calculate_due_date(assignment["module"], assignment["due_day"])
    formatted_assignments.append(f"{assignment['name']} - Due: {due_date.strftime('%Y-%m-%d')}")
    # Adding additional information if BioSig-ID™
    if assignment["type"] == "BioSig-ID™":
        formatted_assignments[-1] += " (BioSig-ID™ validation)"

# Extract dates and sort
sorted_assignments = sorted(formatted_assignments, key=lambda x: datetime.strptime(x.split(" - Due: ")[-1].split(" ")[0], '%Y-%m-%d'))

# Output the sorted list
for assignment in sorted_assignments:
    print(assignment)
```
